```python
# Setup: adiciona backend ao path e carrega .env
import datawork
datawork.setup()
```

## 1. Carregar Dados

Editar o caminho abaixo para apontar para o arquivo da fonte.

```python
from datawork.io.loaders import load_csv

# Opção A: carregar de CSV local
df = load_csv("../data/NOME_DO_ARQUIVO.csv")

# Opção B: carregar da API (fonte já submetida)
# from datawork.io.loaders import load_raw_registros
# df = load_raw_registros("FONTE_ID_AQUI")

print(f"Shape: {df.shape[0]} rows x {df.shape[1]} cols")
print(f"Colunas: {list(df.columns)}")
```

## 2. Profiling Automático

Gera relatório completo com ydata-profiling (pode levar alguns segundos).

```python
from datawork.profiling import profile

report = profile(df, title="Exploração Fonte Nova")
report.to_notebook_iframe()
```

## 3. Completude por Campo

```python
from datawork.profiling import completeness_report

completeness_report(df)
```

## 4. Amostra

```python
from datawork.display import show_sample

show_sample(df, 20, "Primeiras 20 linhas")
```

## 5. Análise por Coluna

Usar as células abaixo para explorar colunas específicas. Adicionar novas células conforme necessário.

```python
from datawork.profiling import value_distribution

# Trocar "NOME_COLUNA" pela coluna que quer analisar
value_distribution(df, "NOME_COLUNA")
```
